```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as sm
import seaborn as sns
import tensorflow as tf # This code has been tested with TensorFlow 1.6
```

```python
#nsebankdata=pd.read_csv('data/IND_NSE/NSEBANK.csv')
AXISBANK=pd.read_csv('data/IND_NSE/AXISBANK.NS.csv')
BANKBARODA=pd.read_csv('data/IND_NSE/BANKBARODA.NS.csv')
FEDERALBNK=pd.read_csv('data/IND_NSE/FEDERALBNK.NS.csv')
HDB=pd.read_csv('data/IND_NSE/HDB.csv')
#IBN=pd.read_csv('data/IND_NSE/IBN.csv')
#IDFCFIRSTB=pd.read_csv('data/IND_NSE/IDFCFIRSTB.csv')
INDUSINDBK=pd.read_csv('data/IND_NSE/INDUSINDBK.NS.csv')
KOTAKBANK=pd.read_csv('data/IND_NSE/KOTAKBANK.NS.csv')
PNB=pd.read_csv('data/IND_NSE/PNB.NS.csv')
#RBLBANK=pd.read_csv('data/IND_NSE/RBLBANK.NS.csv')
SBI=pd.read_csv('data/IND_NSE/SBIN.NS.csv')
YESBANK=pd.read_csv('data/IND_NSE/YESBANK.NS.csv')
```

```python
#nsebankdata['NSEBANK']=nsebankdata['Close']
bankdata=pd.DataFrame(AXISBANK[['Close','Date']])
bankdata['AXISBANK']=AXISBANK['Close']
bankdata['BANKBARODA']=BANKBARODA['Close']
bankdata['FEDERALBNK']=FEDERALBNK['Close']
#bankdata['IBN']=IBN['Close']
#bankdata['IDFCFIRSTB']=IDFCFIRSTB['Close']
bankdata['INDUSINDBK']=INDUSINDBK['Close']
bankdata['KOTAKBANK']=KOTAKBANK['Close']
bankdata['PNB']=PNB['Close']
#bankdata['RBLBANK']=RBLBANK['Close']
bankdata['YESBANK']=YESBANK['Close']
bankdata['SBI']=SBI['Open']
```
